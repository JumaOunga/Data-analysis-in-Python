```python
rows=[]
for i in range(len(netflix)):
    if netflix['date_added'].iloc[i] == "":
        rows.append(i)
```

```python
netflix.loc[rows,:]
```

```python
netflix.date_added
```

```python
month_added=[]
year_added=[]
for i in range(len(netflix)):
    if i in rows:
        month_added.append(0)
        year_added.append(0)
    else:
        date = netflix['date_added'].iloc[i].split(" ")
        month_added.append(date[0])
        year_added.append(date[2])
```

```python
year_added
```

```python
from datetime import datetime
for i, month in enumerate(month_added):
    if month != 0:
        datetime_obj=datetime.strptime(month, "%B")
        month_num = datetime_obj.month
        month_added[i]=month_num
```

```python
print(set(month_added))
print(set(year_added))
```

```python
netflix.insert(7, 'month_added', month_added, allow_duplicates=True)
```

```python
netflix.insert(8, 'year_added', year_added, allow_duplicates= True)
```

```python
shows=[]
films=[]
for i in range(len(netflix)):
    if netflix['type'].iloc[i] == 'TV Show':
        shows.append(i)
    else:
        films.append(i)
```

```python
netflix_shows=netflix.loc[shows,:]
```

```python
netflix_movies = netflix.loc[films,:]
```

```python
netflix_shows.duration.shape
```

```python
netflix_movies.duration.shape
```

```python
netflix_shows = netflix_shows.set_index(pd.Index(range(0,len(netflix_shows))))
netflix_movies = netflix_movies.set_index(pd.Index(range(0,len(netflix_movies))))
```

```python
def getDuration(data):
    duration = []
    for value in data:
        if type(value) is float:
            duration.append(0)
        else:
            values = value.split(" ")
            duration.append(int(values[0]))
    return duration
            
```

```python
netflix_shows.insert(11, 'seasons', getDuration(netflix_shows['duration']))
netflix_shows.drop('duration',axis=1, inplace=True)
```

```python
netflix_movies.insert(11, 'Duration', getDuration(netflix_movies['duration']))
netflix_movies.drop('duration',axis=1, inplace=True)
```
